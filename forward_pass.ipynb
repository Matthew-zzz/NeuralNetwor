#### Простая нейронная сеть без обучения

- Корректировка через Relu.
- Выражение процента через сигмоиду.

---
##### Сначала опишу математические функции(ReLU, sigmoid, скалярное(алгебраическое) произведение векторов), необходимые для дальнейших этапов

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return max(0, x)

def sum_vectors(vector1, vector2):
    sum_v = 0
    for index, elem in enumerate(vector1):
        sum_v += elem * vector2[index][0]

    return round(sum_v, 1).tolist()
```

---

##### Теперь функции слоев 

```python
def hidden_layer(vector, weight):
    return sum_vectors(vector, weight)

def output_layer(hidden_vector, weight):
    return sum_vectors(hidden_vector, weight)
```

##### Готовый макет сети

```python
vector = [1, 0, 1]

hidden_weights1= np.random.uniform(low=-0.5, high=0.5, size=(3, 1))
hidden_weights2= np.random.uniform(low=-0.5, high=0.5, size=(3, 1))
output_weight = np.random.uniform(low=-0.5, high=0.5, size=(2, 1))

def neural_network():
    hidden_vector = [
        relu(hidden_layer(vector, hidden_weights1)), 
        relu(hidden_layer(vector, hidden_weights2))
    ]
    output_vector = sum_vectors(hidden_vector, output_weight)

    print(f"Входной слой: {vector}")
    print(f"Скрытый слой: {hidden_vector}")
    print(f"Выходной слой: {output_vector}")
    print(f"Вероятность спама: {round(sigmoid(output_vector), 3).tolist()}")

neural_network()
```

Первоначальный вид сети без обучения
